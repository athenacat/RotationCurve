```python
from astropy.io import ascii
from astropy.table import Table,QTable
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iso_mini_fits = Table.read('/Users/richardzhang/Documents/UR_stuff/Research_UR/RotationCurve/2D_RC/main/iso_mini_clean.csv',format='ascii.csv')
nfw_mini_fits = Table.read('/Users/richardzhang/Documents/UR_stuff/Research_UR/RotationCurve/2D_RC/main/nfw_mini_clean.csv',format='ascii.csv')
bur_mini_fits = Table.read('/Users/richardzhang/Documents/UR_stuff/Research_UR/RotationCurve/2D_RC/main/bur_mini_clean.csv',format='ascii.csv')
```

```python
iso_mini_fits
```

```python
chi2_norm_iso = iso_mini_fits['chi2']
chi2_norm_nfw = nfw_mini_fits['chi2']
chi2_norm_bur = bur_mini_fits['chi2']
max(chi2_norm_iso),max(chi2_norm_nfw),max(chi2_norm_bur)
```

```python
mass_bins = np.arange(0,2000,50)
```

```python
plt.figure(figsize=(10,8))
plt.hist(chi2_norm_iso,bins=mass_bins,label='$r_{90}$',alpha=0.4);plt.legend();plt.xlim(0,800);plt.xlabel('$\chi^2$');plt.title('Isothermal')
```

```python
plt.figure(figsize=(10,8))
plt.hist(chi2_norm_nfw,bins=mass_bins,label='$r_{90}$',alpha=0.4);plt.xlim(0,800);plt.xlabel('$\chi^2$');plt.title('NFW')
```

```python
plt.figure(figsize=(10,8))
plt.hist(chi2_norm_bur,bins=mass_bins,label='$r_{90}$',alpha=0.4);plt.xlim(0,800);plt.xlabel('$\chi^2$');plt.title('Burket')
```
