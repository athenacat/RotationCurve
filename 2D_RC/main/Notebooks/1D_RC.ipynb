```python
import numpy as np

import matplotlib.pyplot as plt

from astropy.table import QTable

from astropy.io import ascii

from scipy.optimize import minimize

import math

import os.path
from os import path
from galaxy_component_functions import vel_tot_iso,\
                                       vel_tot_NFW,\
                                       vel_tot_bur
```

```python
RC_FILE_FOLDER = '/Users/richardzhang/Documents/UR_Stuff/Research_UR/data/DRP-rot_curve_data_files/'
```

```python
galaxy_ID = []
plate = ['7443','7495','7815']
IFU = ['1901','1902','3701','3702','3703','3704','6101','6102','6103','6104','9101','9102','12701','12702','12703','12704','12705']
for i in range(len(plate)):
    for j in range(len(IFU)):
        galaxy_ID.append(plate[i]+'-'+IFU[j])
```

```python
for i in range(len(galaxy_ID)):
    rot_curve_file = RC_FILE_FOLDER + galaxy_ID[i] + '_rot_curve_data.txt'
    if path.exists(rot_curve_file):
        rc_file = QTable.read(rot_curve_file, format='ascii.ecsv')
        r_dep = rc_file['deprojected_distance'].data
        av = rc_file['rot_vel_avg'].data
        plt.plot(r_dep, av)
        plt.show()
```
