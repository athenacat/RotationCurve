```python
# Isothermal

iso_MCMC_mini_corr_ID = []
iso_MCMC_mini_corr_rho0_b = []
iso_MCMC_mini_corr_Rb = []
iso_MCMC_mini_corr_SigD = []
iso_MCMC_mini_corr_Rd = []
iso_MCMC_mini_corr_rho0_h = []
iso_MCMC_mini_corr_Rh = []
iso_MCMC_mini_corr_incl = []
iso_MCMC_mini_corr_phi = []
iso_MCMC_mini_corr_xcen = []
iso_MCMC_mini_corr_ycen = []
iso_MCMC_mini_corr_Vsys = []
iso_MCMC_mini_corr_chi2 = []

for i in range(31):
    dtable = Table.read('/Users/richardzhang/Documents/UR_stuff/Research_UR/RotationCurve/2D_RC/main/minimize_mcmc_corr_iso/iso_mcmc_corr_mini_b'+str(i+1)+'.csv',format='ascii.csv')
    for j in range(len(dtable)):
        if np.isfinite(dtable['chi2'][j]):
            iso_MCMC_mini_corr_ID.append(dtable['gal_ID'][j])
            iso_MCMC_mini_corr_rho0_b.append(dtable['rho0_b'][j])
            iso_MCMC_mini_corr_Rb.append(dtable['Rb'][j])
            iso_MCMC_mini_corr_SigD.append(dtable['SigD'][j])
            iso_MCMC_mini_corr_Rd.append(dtable['Rd'][j])
            iso_MCMC_mini_corr_rho0_h.append(dtable['rho0_h'][j])
            iso_MCMC_mini_corr_Rh.append(dtable['Rh'][j])
            iso_MCMC_mini_corr_incl.append(dtable['incl'][j])
            iso_MCMC_mini_corr_phi.append(dtable['phi'][j])
            iso_MCMC_mini_corr_xcen.append(dtable['x_cen'][j])
            iso_MCMC_mini_corr_ycen.append(dtable['y_cen'][j])
            iso_MCMC_mini_corr_Vsys.append(dtable['Vsys'][j])
            iso_MCMC_mini_corr_chi2.append(dtable['chi2'][j])
        
# NFW 

nfw_MCMC_mini_corr_ID = []
nfw_MCMC_mini_corr_rho0_b = []
nfw_MCMC_mini_corr_Rb = []
nfw_MCMC_mini_corr_SigD = []
nfw_MCMC_mini_corr_Rd = []
nfw_MCMC_mini_corr_rho0_h = []
nfw_MCMC_mini_corr_Rh = []
nfw_MCMC_mini_corr_incl = []
nfw_MCMC_mini_corr_phi = []
nfw_MCMC_mini_corr_xcen = []
nfw_MCMC_mini_corr_ycen = []
nfw_MCMC_mini_corr_Vsys = []
nfw_MCMC_mini_corr_chi2 = []

for k in range(32):
    dtable = Table.read('/Users/richardzhang/Documents/UR_stuff/Research_UR/RotationCurve/2D_RC/main/minimize_mcmc_corr_nfw/nfw_mcmc_corr_mini_b'+str(k+1)+'.csv',format='ascii.csv')
    for l in range(len(dtable)):
        if np.isfinite(dtable['chi2'][l]):
            nfw_MCMC_mini_corr_ID.append(dtable['gal_ID'][l])
            nfw_MCMC_mini_corr_rho0_b.append(dtable['rho0_b'][l])
            nfw_MCMC_mini_corr_Rb.append(dtable['Rb'][l])
            nfw_MCMC_mini_corr_SigD.append(dtable['SigD'][l])
            nfw_MCMC_mini_corr_Rd.append(dtable['Rd'][l])
            nfw_MCMC_mini_corr_rho0_h.append(dtable['rho0_h'][l])
            nfw_MCMC_mini_corr_Rh.append(dtable['Rh'][l])
            nfw_MCMC_mini_corr_incl.append(dtable['incl'][l])
            nfw_MCMC_mini_corr_phi.append(dtable['phi'][l])
            nfw_MCMC_mini_corr_xcen.append(dtable['x_cen'][l])
            nfw_MCMC_mini_corr_ycen.append(dtable['y_cen'][l])
            nfw_MCMC_mini_corr_Vsys.append(dtable['Vsys'][l])
            nfw_MCMC_mini_corr_chi2.append(dtable['chi2'][l])
            
# Burkert

bur_MCMC_mini_corr_ID = []
bur_MCMC_mini_corr_rho0_b = []
bur_MCMC_mini_corr_Rb = []
bur_MCMC_mini_corr_SigD = []
bur_MCMC_mini_corr_Rd = []
bur_MCMC_mini_corr_rho0_h = []
bur_MCMC_mini_corr_Rh = []
bur_MCMC_mini_corr_incl = []
bur_MCMC_mini_corr_phi = []
bur_MCMC_mini_corr_xcen = []
bur_MCMC_mini_corr_ycen = []
bur_MCMC_mini_corr_Vsys = []
bur_MCMC_mini_corr_chi2 = []

for s in range(30):
    dtable = Table.read('/Users/richardzhang/Documents/UR_stuff/Research_UR/RotationCurve/2D_RC/main/minimize_mcmc_corr_bur/bur_mcmc_corr_mini_b'+str(s+1)+'.csv',format='ascii.csv')
    for t in range(len(dtable)):
        if np.isfinite(dtable['chi2'][t]):
            bur_MCMC_mini_corr_ID.append(dtable['gal_ID'][t])
            bur_MCMC_mini_corr_rho0_b.append(dtable['rho0_b'][t])
            bur_MCMC_mini_corr_Rb.append(dtable['Rb'][t])
            bur_MCMC_mini_corr_SigD.append(dtable['SigD'][t])
            bur_MCMC_mini_corr_Rd.append(dtable['Rd'][t])
            bur_MCMC_mini_corr_rho0_h.append(dtable['rho0_h'][t])
            bur_MCMC_mini_corr_Rh.append(dtable['Rh'][t])
            bur_MCMC_mini_corr_incl.append(dtable['incl'][t])
            bur_MCMC_mini_corr_phi.append(dtable['phi'][t])
            bur_MCMC_mini_corr_xcen.append(dtable['x_cen'][t])
            bur_MCMC_mini_corr_ycen.append(dtable['y_cen'][t])
            bur_MCMC_mini_corr_Vsys.append(dtable['Vsys'][t])
            bur_MCMC_mini_corr_chi2.append(dtable['chi2'][t])
```

```python
improve_iso = []
lost_iso = []
for o in range(len(iso_mini_fits)):
    for p in range(len(iso_MCMC_mini_corr_ID)):
        if iso_mini_fits['galaxy_ID'][o] == iso_MCMC_mini_corr_ID[p]:
            if iso_mini_fits['chi2'][o] > iso_MCMC_mini_corr_chi2[p]:
                improve_iso.append(iso_mini_fits['galaxy_ID'][o])
            elif iso_MCMC_mini_corr_chi2[p] > 150:
                lost_iso.append(iso_mini_fits['galaxy_ID'][o])
            iso_mini_fits['rho0_b'][o] = iso_MCMC_mini_corr_rho0_b[p]
            iso_mini_fits['Rb'][o] = iso_MCMC_mini_corr_Rb[p]
            iso_mini_fits['SigD'][o] = iso_MCMC_mini_corr_SigD[p]
            iso_mini_fits['Rd'][o] = iso_MCMC_mini_corr_Rd[p]
            iso_mini_fits['rho0_h'][o] = iso_MCMC_mini_corr_rho0_h[p]
            iso_mini_fits['Rh'][o] = iso_MCMC_mini_corr_Rh[p]
            iso_mini_fits['incl'][o] = iso_MCMC_mini_corr_incl[p]
            iso_mini_fits['phi'][o] = iso_MCMC_mini_corr_phi[p]
            iso_mini_fits['x_cen'][o] = iso_MCMC_mini_corr_xcen[p]
            iso_mini_fits['y_cen'][o] = iso_MCMC_mini_corr_ycen[p]
            iso_mini_fits['Vsys'][o] = iso_MCMC_mini_corr_Vsys[p]
            iso_mini_fits['chi2'][o] = iso_MCMC_mini_corr_chi2[p]

improve_nfw = []
lost_nfw = []
for a in range(len(nfw_mini_fits)):
    for b in range(len(nfw_MCMC_mini_corr_ID)):
        if nfw_mini_fits['galaxy_ID'][a] == nfw_MCMC_mini_corr_ID[b]:
            if nfw_mini_fits['chi2'][a] > nfw_MCMC_mini_corr_chi2[b]:
                improve_nfw.append(nfw_mini_fits['galaxy_ID'][a])
            elif nfw_MCMC_mini_corr_chi2[b] > 150:
                lost_nfw.append(nfw_mini_fits['galaxy_ID'][a])
            nfw_mini_fits['rho0_b'][a] = nfw_MCMC_mini_corr_rho0_b[b]
            nfw_mini_fits['Rb'][a] = nfw_MCMC_mini_corr_Rb[b]
            nfw_mini_fits['SigD'][a] = nfw_MCMC_mini_corr_SigD[b]
            nfw_mini_fits['Rd'][a] = nfw_MCMC_mini_corr_Rd[b]
            nfw_mini_fits['rho0_h'][a] = nfw_MCMC_mini_corr_rho0_h[b]
            nfw_mini_fits['Rh'][a] = nfw_MCMC_mini_corr_Rh[b]
            nfw_mini_fits['incl'][a] = nfw_MCMC_mini_corr_incl[b]
            nfw_mini_fits['phi'][a] = nfw_MCMC_mini_corr_phi[b]
            nfw_mini_fits['x_cen'][a] = nfw_MCMC_mini_corr_xcen[b]
            nfw_mini_fits['y_cen'][a] = nfw_MCMC_mini_corr_ycen[b]
            nfw_mini_fits['Vsys'][a] = nfw_MCMC_mini_corr_Vsys[b]
            nfw_mini_fits['chi2'][a] = nfw_MCMC_mini_corr_chi2[b]

improve_bur = []
lost_bur = []
for c in range(len(bur_mini_fits)):
    for d in range(len(bur_MCMC_mini_corr_ID)):
        if bur_mini_fits['galaxy_ID'][c] == bur_MCMC_mini_corr_ID[d]:
            if bur_mini_fits['chi2'][c] > bur_MCMC_mini_corr_chi2[d]:
                improve_bur.append(bur_mini_fits['galaxy_ID'][c])
            elif bur_MCMC_mini_corr_chi2[d] > 150:
                lost_bur.append(bur_mini_fits['galaxy_ID'][c])
            bur_mini_fits['rho0_b'][c] = bur_MCMC_mini_corr_rho0_b[d]
            bur_mini_fits['Rb'][c] = bur_MCMC_mini_corr_Rb[d]
            bur_mini_fits['SigD'][c] = bur_MCMC_mini_corr_SigD[d]
            bur_mini_fits['Rd'][c] = bur_MCMC_mini_corr_Rd[d]
            bur_mini_fits['rho0_h'][c] = bur_MCMC_mini_corr_rho0_h[d]
            bur_mini_fits['Rh'][c] = bur_MCMC_mini_corr_Rh[d]
            bur_mini_fits['incl'][c] = bur_MCMC_mini_corr_incl[d]
            bur_mini_fits['phi'][c] = bur_MCMC_mini_corr_phi[d]
            bur_mini_fits['x_cen'][c] = bur_MCMC_mini_corr_xcen[d]
            bur_mini_fits['y_cen'][c] = bur_MCMC_mini_corr_ycen[d]
            bur_mini_fits['Vsys'][c] = bur_MCMC_mini_corr_Vsys[d]
            bur_mini_fits['chi2'][c] = bur_MCMC_mini_corr_chi2[d]
```
